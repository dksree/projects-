```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import pytz
```

```python
df = pd.read_csv("google_play_store_dataset.csv")
df.head()
```

```python
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')

df['Size'] = df['Size'].str.replace('M','')
df['Size'] = pd.to_numeric(df['Size'], errors='coerce')

df['Last Updated'] = pd.to_datetime(df['Last Updated'], errors='coerce')
```

```python
df = df[(df['Rating'] >= 4.0) & (df['Size'] >= 10) & (df['Last Updated'].dt.month == 1)]
```

```python
df['Installs'] = df['Installs'].str.replace('[+,]', '', regex=True)
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')
```

```python
top_categories = df.groupby('Category')['Installs'].sum().sort_values(ascending=False).head(10)

df = df[df['Category'].isin(top_categories.index)]
```

```python
result = df.groupby('Category').agg({
    'Rating':'mean',
    'Reviews':'sum'
}).reset_index()

result
```

```python
ist = pytz.timezone('Asia/Kolkata')
current_time = datetime.now(ist).time()

start = datetime.strptime("15:00","%H:%M").time()
end = datetime.strptime("17:00","%H:%M").time()
```

```python
if start <= current_time <= end:

    x = range(len(result))

    plt.figure(figsize=(12,6))

    plt.bar(x, result['Rating'], width=0.4, label='Average Rating')
    plt.bar([i+0.4 for i in x], result['Reviews'], width=0.4, label='Total Reviews')

    plt.xticks([i+0.2 for i in x], result['Category'], rotation=45)

    plt.title("Average Rating vs Total Reviews for Top 10 Categories")
    plt.xlabel("Category")
    plt.ylabel("Values")

    plt.legend()

    plt.show()

else:
    print("Graph available only between 3 PM and 5 PM IST")
```
